La función apriori tiene un dataframe en formato booleano como input, por lo que habrá que transformar el df. En un dataframe booleano las filas representan un pedido y las columnas los distintos productos. Si en la fila i, columna j, hay una True, significa que el pedido i se incluyó el producto j. Para hacer esta transformación pasaremos el df a una lista de listas ya que en la documentación de apriori() incluyen un código que crea un df booleano en base a una lista de listas.

```python
lista_og = df.values.tolist()
```

```python
lista_limpia = []

for i in range(len(lista_og)):
    # Para Cada fila seleccionamos los no Na
    filtered_iterator = filter(lambda x: pd.notna(x), lista_og[i])

    # Convertimos esto en una lista
    lista_limpia.append(list(filtered_iterator))
#Comprobamos que el primer pedido ha sido guardado en la lista correctamente
print(lista_limpia[0])
```

```python
#Nos basamos en el ejemplo dado en la documentación de Apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth


dataset = lista_limpia
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Limpiamos los datos
df = df.dropna()  # Eliminamos filas con valores nulos
```

```python
#Para simplificar el algoritmo de extracción de secuencias 
#ordenaremos las filas por clientes y para cada cliente se ordenarán por el ID de pedido. 
#Habrá una secuenica de transacciones por cada usuario, por lo que ayuda tener los datos agrupados.
df.sort_values(['CustomerID', 'InvoiceNo'])      
df = df.reset_index(drop=True)     
```

```python
#Versión lista de listas de tuplas: una secuencia <(ab)ac> =  [(ab), (a), (c)]
num_users = len(pd.unique(df['CustomerID']))
num_seq = 0
transactions = []
index = 0
while(num_seq<num_users):
    user = df.iloc[index, 6]
    invoice_num = df.iloc[index, 0]
    sequence = []
    while(df.iloc[index, 6] == user):
        subsequence = ()


        while(df.iloc[index, 0] == invoice_num):
            subsequence = subsequence + (df.iloc[index, 2],)
            index = index + 1
        #print(subsequence)
        invoice_num = df.iloc[index, 0]
        sequence.append(subsequence)

    transactions.append(sequence)
    num_seq = num_seq + 1
    user = df.iloc[index, 6]
print(transactions[0])
```
